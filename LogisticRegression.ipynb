```python
def get_x_y(data):
    a=data.copy()
    def f(s):
        if s=="female":
            return 0
        return 1
    a["sex"]=a.Sex.apply(f)
    def f1(s):
        s=str(s).strip()
        if s=="S":
            return 1
        elif s=="C":
            return 0
        return -1
    a["sex"]=a.Sex.apply(f)
    a["embark"]=a.Embarked.apply(f1)
    survived=a["Survived"]
    # embarked s,c,q
    a=a.drop(["Name","Survived","Sex","Ticket","Cabin","Embarked"], axis='columns')
    a["Survived"]=survived;
    m=round(a[a.Survived==1].Age.mean(),2)
#     a[a.Survived==1].Age.fillna(m,inplace=True)
#     m=round(a[a.Survived==0].Age.mean(),2)
#     a[a.Survived==0].Age.fillna(m,inplace=True)
#     m=round(a.Age.mean(),2)
    a.Age.fillna(m,inplace=True)
    
    a.dropna(inplace=True)
    d=pd.DataFrame(a).to_numpy()
    x=d[:,:d.shape[1]-1]
    # print(d)
    # print(x)

    y=d[:,d.shape[1]-1]
    # print(y)
    return x,y
```

```python
import numpy as np
def h(x):
    import math
    e=math.e
    den=1+math.pow(e,-1*x)
    return 1/den
def step_gradient(points,y_data,learning_rate,coef_array):
    slope_array=[0]*(len(coef_array))
    N=len(points)
    new_slope_array=[0]*(len(coef_array))
    for i in range(N):
        
        n=points.shape[1]
        s=0
        for k in range(n):
            s+=coef_array[k]*points[i,k]
        s=h(s)
        y=y_data[i]
        for j in range(n):

            x=points[i,j]
#             slope_array[j]=slope_array[j]+(((-2/N)*(y-coef_array[j]*x-coef_array[n-1]))*x)
#             slope_array[j]=slope_array[j]+(((-2/N)*(y-s)*x))
            slope_array[j]=(((-1/N)*(y-s)*x))
        for j in range(n):
            new_slope_array[j]=coef_array[j]-learning_rate*slope_array[j]
#     n=points.shape[1]        
    
    return new_slope_array
def gd(points,y_data,num_features,learning_rate,num_iterations):
    coef_array=[0]*(num_features)
    for i in range(num_iterations):
        coef_array=step_gradient(points,y_data,learning_rate,coef_array)
#         print(i,cost(points,y_data,coef_array))
        
#     print("coef array",coef_array)
#     print(len(coef_array)," ------------- array len")
    return coef_array
# def cost(points,y_data,coef_array):
#     total_cost=0
#     M=len(points)
#     for i in range(M):
#         n=points.shape[1]
#         y=y_data[i]
#         sum=0
#         for j in range(n):
#             x=points[i,j]
#             sum+=coef_array[j]*x
#         y=points[i,1]
#         total_cost+=(1/M)*((y-sum)**2)
#     return total_cost
def predict(x_data,coef_array):

    x_data=np.array(x_data)
    coef_array=np.array(coef_array)
    
    for i in range(x_data.shape[0]):
        x_data[i,:]=x_data[i,:]*coef_array
    y_data=np.sum(x_data,axis=1)
    new_y=[]
    print("===========================")
    print(y_data.shape)
    print(y_data)
    for i in range(y_data.shape[0]):
        print(i,h(y_data[i]) )
        if(h(y_data[i])>0.5):
            new_y.append(1)
        else:
            new_y.append(0)
        
    return new_y
def score(y_truth,y_pred):
#     print("--------")
    y_truth=np.array(y_truth)
    y_test=np.array(y_pred)
    y=y_truth-y_test
    count_0=0
    for i in y:
        if(i==1):
            count_0=count_0+1;
    
    return float(count_0)/float(y.shape[0]) 
```
